## SETTING MAP FUNCTIONS

```python
def map_stadium(stadium):
    stadium_map = {
        'Sheikh Zayed': 0,
        'Dubai': 1,
        'Sharjah': 2
    }
    return stadium_map.get(stadium, -1)
all_rounders['Ground'] = all_rounders['Ground'].map(map_stadium)
batsmen['Ground'] = batsmen['Ground'].map(map_stadium)
bowlers['Ground'] = bowlers['Ground'].map(map_stadium)
```

```python
def map_match_type(match_type):
    if match_type == 'Defending':
        return 0
    else:
        return 1

# Apply mapping function to 'Match_Type' column in all DataFrames
all_rounders['Match_Type'] = all_rounders['Match_Type'].map(map_match_type)
batsmen['Match_Type'] = batsmen['Match_Type'].map(map_match_type)
bowlers['Match_Type'] = bowlers['Match_Type'].map(map_match_type)
```

```python
def hand(hand):
    if hand == 'RH':
        return 0
    else:
        return 1

# Apply mapping function to 'Match_Type' column in all DataFrames
all_rounders['RH/LH'] = all_rounders['RH/LH'].map(hand)
batsmen['RH/LH'] = batsmen['RH/LH'].map(hand)
bowlers['RH/LH'] = bowlers['RH/LH'].map(hand)
```

```python
def map_opposite_team(team):
    if team == 'CSK':
        return 7
    elif team == 'DC':
        return 2
    elif team == 'KKR':
        return 5
    elif team == 'KXIP':
        return 6
    elif team == 'MI':
        return 1
    elif team == 'RCB':
        return 4
    elif team == 'RR':
        return 8
    elif team == 'SRH':
        return 3
    else:
        return -1  # Unknown team

# Apply mapping function to 'Opp_Team' column in all DataFrames
all_rounders['Opp_Team'] = all_rounders['Opp_Team'].map(map_opposite_team)
batsmen['Opp_Team'] = batsmen['Opp_Team'].map(map_opposite_team)
bowlers['Opp_Team'] = bowlers['Opp_Team'].map(map_opposite_team)
```

```python
# Loop through each role group and calculate predicted points using the trained models
import pandas as pd

def dreamteam(players_df,Team1,Team2):
    # Separate players from DC and CSK
    dc_players = players_df[players_df['Team'] == Team1]
    csk_players = players_df[players_df['Team'] == Team2]

    # Select top 4 players from each team
    dc_top_4 = dc_players.nlargest(4, 'Predicted_Points')
    csk_top_4 = csk_players.nlargest(4, 'Predicted_Points')

    # Concatenate the top players from both teams
    selected_team = pd.concat([dc_top_4, csk_top_4])

    # Check if a wicket-keeper is already selected
    if 'WK' not in selected_team['Role'].values:
        # Select the best wicket-keeper available
        best_wk = players_df[players_df['Role'] == 'WK'].nlargest(1, 'Predicted_Points')
        selected_team = pd.concat([selected_team, best_wk])

    # Check if there are at least 2 bowlers in the selected team
    num_bowlers = (selected_team['Role'] == 'BALL').sum()
    if num_bowlers < 2:
        # Select the best available bowlers to complete the team
        num_bowlers_needed = 2 - num_bowlers
        best_bowlers = players_df[~players_df.index.isin(selected_team.index) & (players_df['Role'] == 'BALL')].nlargest(num_bowlers_needed, 'Predicted_Points')
        selected_team = pd.concat([selected_team, best_bowlers])

    # Check if the team has exactly 11 players
    if len(selected_team) < 11:
        # Select the best remaining players to complete the team
        num_players_needed = 11 - len(selected_team)
        best_remaining = players_df[~players_df.index.isin(selected_team.index)].nlargest(num_players_needed, 'Predicted_Points')
        selected_team = pd.concat([selected_team, best_remaining])

    # Reset the index to 1-based indexing
    selected_team.index = range(1, len(selected_team) + 1)

    # Double points for the highest predicted points
    selected_team.loc[1, 'Predicted_Points'] *= 2

    # 1.5 times points for the second-highest predicted points
    selected_team.loc[2, 'Predicted_Points'] *= 1.5

    # Add (C) and (VC) to the names
    selected_team.loc[1, 'Player'] = selected_team.loc[1, 'Player'] + ' (C)'
    selected_team.loc[2, 'Player'] = selected_team.loc[2, 'Player'] + ' (VC)'

    return selected_team


# Now you can use the final_df DataFrame for further processing and selecting the best team.

# Assuming you have defined the 'teams' DataFrame and the 'ranking' DataFrame
import pandas as pd
from sklearn.impute import SimpleImputer

# ... (Define your data, models, and functions here)

# Now modify the 'f' function to handle missing values
def f(role_grouped, final_df, Team1, Team2):
    for role, role_group_df in role_grouped:
        # Prepare the features (X) for prediction for the current role group
        columns_order = ['Player', 'Opp_Team', 'RH/LH', 'Match_Type', 'Ground']  # Specify the desired order of columns
        X_test = role_group_df.drop(['Player', 'Role', 'Rank', 'Stadium'], axis=1)  # Remove 'Predicted_Points' from drop list
        X_test['Ground'] = role_group_df['Stadium']  # Include 'Stadium' feature in X_test
        X_test['Player'] = role_group_df['Rank']  # Include 'Rank' feature in X_test
        X_test = X_test.reindex(columns=columns_order)  # Reorder the columns

        # Create a SimpleImputer to fill missing values with the mean of the feature
        imputer = SimpleImputer(strategy='mean')

        # Fill missing values in the test data
        X_test_filled = pd.DataFrame(imputer.fit_transform(X_test), columns=X_test.columns)

        # Scale the features
        X_test_scaled = scaler.transform(X_test_filled)

        if role == 'ALL':
            # Use the trained models to make predictions on the test data
            role_group_df['Predicted_Points'] = all_rounders_linear_regression.predict(X_test_scaled)
        elif role == 'BAT' or role == 'WK':
            role_group_df['Predicted_Points'] = batsmen_linear_regression.predict(X_test_scaled)
        else:
            role_group_df['Predicted_Points'] = bowlers_linear_regression.predict(X_test_scaled)

        # Reset the index to remove serial numbers before storing in final_df
        role_group_df.reset_index(drop=True, inplace=True)

        # Keep only 'Player', 'Team', 'Role', and 'Predicted_Points' columns
        role_group_df = role_group_df[['Player', 'Team', 'Role', 'Predicted_Points']]

        # Concatenate the current role_group_df to final_df
        final_df = pd.concat([final_df, role_group_df], ignore_index=True)

        # Format and display the results for the current role
        for _, row in role_group_df.iterrows():
            player = row['Player'].ljust(20)
            team = row['Team'].ljust(10)
            role = row['Role'].ljust(10)
            predicted_points = str(row['Predicted_Points'])
            print(f"{player}{team}{role}{predicted_points}")
   
        
    selected_team = dreamteam(final_df,Team1,Team2)
       
    # Print the results
    if selected_team is not None:
        print()
        print()
        print()
        print()
        print("Selected Team:")
        print(selected_team[['Player', 'Role', 'Team', 'Predicted_Points']])
        print()
        print("Total Predicted Points: ", selected_team['Predicted_Points'].sum())

# Now you can use the final_df DataFrame for further processing and selecting the best team.

# Assuming you have defined the 'teams' DataFrame and the 'ranking' DataFrame

# Call the function with the provided arguments
def select_team(Team1, Team2, stadium, toss_result, teams, ranking, hand):
    team1_df = teams[teams['Team'] == Team1].copy()
    team2_df = teams[teams['Team'] == Team2].copy()

    # Drop the 'Team_ID' column from both Team1 and Team2 DataFrames
    team1_df.drop('Team_ID', axis=1, inplace=True)
    team2_df.drop('Team_ID', axis=1, inplace=True)

    # Set the 'Opp_Team' map value as the opposite team for every player in Team1 and Team2
    team1_df['Opp_Team'] = map_opposite_team(Team2)
    team2_df['Opp_Team'] = map_opposite_team(Team1)

    # Set the 'Stadium' map value as the ground value for every player in Team1 and Team2
    team1_df['Stadium'] = map_stadium(stadium)
    team2_df['Stadium'] = map_stadium(stadium)

    # Set the 'match_type' column based on the 'toss_result' for both Team1 and Team2
    team1_df['Match_Type'] = toss_result
    team2_df['Match_Type'] = toss_result

    # Assuming 'ranking' DataFrame is available with a mapping of players to their ranks
    team1_df['Rank'] = team1_df['Player'].map(ranking.set_index('Player')['Rank'])
    team2_df['Rank'] = team2_df['Player'].map(ranking.set_index('Player')['Rank'])

    # Map the 'RH/LH' column for both teams using the 'hand' function
    team1_df['RH/LH'] = team1_df['RH/LH'].map(hand)
    team2_df['RH/LH'] = team2_df['RH/LH'].map(hand)

    # Define the desired column order
    desired_columns = ['Player', 'Team', 'Role', 'Rank', 'Opp_Team', 'RH/LH', 'Match_Type', 'Stadium']

    # Reindex both DataFrames with the desired column order
    team1_df = team1_df.reindex(columns=desired_columns)
    team2_df = team2_df.reindex(columns=desired_columns)

    # Concatenate the two DataFrames vertically
    combined_df = pd.concat([team1_df, team2_df])
    final_df = pd.DataFrame()
    role_grouped = combined_df.groupby('Role')
    f(role_grouped, final_df, Team1, Team2)

# Now you can call the select_team function with the provided arguments
select_team('MI', 'KXIP', 'Dubai', 1, teams, ranking, hand)
```
